```python
import pandas as pd

FILES_PATH = "../data/"
```

```python
#DECISION TREE
DT_1 = pd.read_json(FILES_PATH + "DT_METRICAS_1.json")
DT_1_FS = pd.read_json(FILES_PATH + "DT_METRICAS_1_fs.json")
DT_2 = pd.read_json(FILES_PATH + "DT_METRICAS_2.json")
DT_2_FS = pd.read_json(FILES_PATH + "DT_METRICAS_2_fs.json")

#RANDOM FOREST
RF_1 = pd.read_json(FILES_PATH + "RF_METRICAS_1.json")
RF_1_FS = pd.read_json(FILES_PATH + "RF_METRICAS_1_fs.json")
RF_2 = pd.read_json(FILES_PATH + "RF_METRICAS_2.json")
RF_2_FS = pd.read_json(FILES_PATH + "RF_METRICAS_2_fs.json")

#LINEAR REGRESSION
LR_1 = pd.read_json(FILES_PATH + "LR_METRICAS_1.json")
LR_1_FS = pd.read_json(FILES_PATH + "LR_METRICAS_1_fs.json")
LR_2 = pd.read_json(FILES_PATH + "LR_METRICAS_2.json")
LR_2_FS = pd.read_json(FILES_PATH + "LR_METRICAS_2_fs.json")

#LOGISTIC REGRESSION
GR_1 = pd.read_json(FILES_PATH + "GR_METRICAS_1.json")
GR_1_FS = pd.read_json(FILES_PATH + "GR_METRICAS_1_fs.json")
GR_2 = pd.read_json(FILES_PATH + "GR_METRICAS_2.json")
GR_2_FS = pd.read_json(FILES_PATH + "GR_METRICAS_2_fs.json")

#SUPPORT VECTOR MACHINE 
SVM_1 = pd.read_json(FILES_PATH + "SVM_METRICAS_1.json")
SVM_1_FS = pd.read_json(FILES_PATH + "SVM_METRICAS_1_fs.json")
SVM_2 = pd.read_json(FILES_PATH + "SVM_METRICAS_2.json")
SVM_2_FS = pd.read_json(FILES_PATH + "SVM_METRICAS_2_fs.json")



```

```python
new_df = pd.DataFrame([DT_1.loc["100%"], RF_1.loc["100%"], LR_1.loc["100%"], GR_1.loc["100%"], SVM_1.loc["100%"]], index=["DT", "RF", "LR", "GR", "SVM"])

final_df = []
for index in new_df.index:
    for i in new_df.loc[index].index:
        row = {
            "Algoritmos" : index,
            "Métricas" : i,
            "Valores" : new_df[i][index]
        }
        final_df.append(row)
        del row


new_df = pd.json_normalize(final_df)
```
